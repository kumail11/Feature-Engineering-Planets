```python
planets_df.dropna(axis = 0, subset = 'name', inplace = True)
```

```python
planets_df = planets_df.rename(columns = {'name': 'planet_name'})
```

```python
planets_df.head()
```

```python
planets_df['planet_name'].unique()
```

```python
planets_df['terrain'] = planets_df['terrain'].str.split(', ')
```

```python
planets_df['terrain'].head()
```

```python
terrain_bridge = planets_df.explode('terrain')
terrain_bridge
```

```python
terrain_bridge = terrain_bridge.loc[:, ['planet_name', 'terrain']]
terrain_bridge.reset_index(drop = True, inplace = True)
terrain_bridge
```

```python
terrain_bridge = terrain_bridge.dropna(axis = 0, subset = 'terrain')
terrain_bridge
```

```python
terrain_list = list(terrain_bridge['terrain'].unique())
terrain_list.sort()
```

```python
terrain_df = pd.DataFrame.from_dict({'terrain': terrain_list})
terrain_df
```

```python
for terrain in terrain_list:
    plural_terrain = f"{terrain}s"
    if plural_terrain in terrain_list:
        print(f"{terrain} and {plural_terrain}")
```
